```python
from skspatial.objects import Line, Cylinder
from skspatial.plotting import plot_3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
Lz = 4
Lx = 2

cylinder = Cylinder([0, 0, -Lz/2], [0, 0, Lz], Lx/2)
```

## Randomly generate N neutrons

```python
zi = np.random.rand(1)[0] * Lz
xi = np.random.rand(1)[0] * Lx
ki = 5.8

line = Line([-1, 0, 0], [1, 0, 0])
```

```python
line = Line([-1, 0, 0], [1, 0, 0])

A, B = cylinder.intersect_line(line, infinite=False)


plot_3d(
    line.plotter(c='k'),
    cylinder.plotter(alpha=0.2),
    A.plotter(c='r', s=100),
    B.plotter(c='r', s=100),
)
```

```python
f = lambda x: np.exp(-x**2/2)
g = lambda x: np.exp(-x**4/2)


X = np.linspace(-4,4,100)

F = f(X)
F /=F.sum()



out1 = np.random.choice(np.arange(100), p=F, size = 10000)
plt.hist(X[out1], density=True)
plt.plot(X,F, c='red')


G = g(X)
G /=G.sum()
plt.plot(X,G, c='red')
plt.plot(X**2,G*(2*X), c='red')

```
