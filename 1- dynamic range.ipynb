# Dynamic range
From the momentum and energy conservation
$$
\begin{aligned}
\vec{Q} = \vec{k}_i - \vec{k}_f \\
\omega = \omega_i - \omega_f
\end{aligned}
$$
and the vectorial relation
$$
|\vec{k}_i - \vec{k}_f| = \sqrt{k_i^2 + k_f^2 - 2 k_i k_f \cos \theta}
$$
we can derive the dynamic range
$$
Q = k_i \sqrt{2 + \frac{\omega}{E_i} + 2\sqrt{1 + \frac{\omega}{E_i}}\cos\theta} \hspace{1cm} k_i=2\pi\sqrt{\frac{E_i}{81.80}}
$$
In our case the role played by the scattering angle $\theta$ is to limit the dynamic range because of the direct beam contribution. In the experiment this angle is limited to $\theta > 1^\circ$.


```python
Q = lambda omega, theta, Ei: 2*np.pi*np.sqrt(Ei/81.80) * np.sqrt(2 - omega/Ei - 2*np.sqrt(1 - omega/Ei)*np.cos(theta))
```

```python
Ef = 74.5

omega = np.arange (-100, 100, 0.01)

plt.figure(figsize=(10, 5))
plt.plot(omega, Q(omega, np.deg2rad(0), Ef), label='Ei = '+str(74.5)+' meV', color='red')
plt.plot(omega, Q(omega, np.deg2rad(180), Ef), label='Ei = '+str(74.5)+' meV', color='blue')



plt.xlabel('Energy [meV]')
plt.ylabel('Q [$\AA^{-1}$]')
plt.xlim(-100,100)
plt.ylim(0, 15)
plt.legend()
```
